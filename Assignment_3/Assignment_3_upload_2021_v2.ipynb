#### 4. Write a function that prints any word that appeared in the last 20% of a text that had not been encountered earlier. Use text1 from nltk.book to call this function.

```python
text1 = nltk.book.text1
```

```python
def word_20(text1):
    cut = int(.8 * len(text1))
    for w in text1[cut:]:
        if w not in text1[:cut]:
            print(w)

word_20(text1)
```

#### 5. Write a program that takes the sentence ("we have seen two kinds of two sequence objects") expressed as a single string, splits it and counts up the tokens. Get it to print out each token and the token's frequency, one per line, in alphabetical order. You should write a function and call that function to process the sentence.

```python
sent_5 = "we have seen two kinds of two sequence objects"
```

```python
def token_counter(raw):
    text = word_tokenize(raw)
    fdist = nltk.FreqDist(text)
    for key in sorted(fdist):
        print((key) + ':', fdist[key])


token_counter(sent_5)
```

#### 6. Write a function shorten(word, n) to process a text (big big big world today tomorrow good Today good), omitting the n most frequently occurring words of the text. You should use w.lower() to normalize the text first. Please call this shorten function.

```python
sent_6 = 'big big big world today tomorrow good Today good'
```

```python
def shorten(sent, n):
    text = sent.split()
    most_freq = set(w for (w,f) in nltk.FreqDist(text).most_common(n))
    return sorted([w.lower() for w in text if w not in most_freq])

shorten(sent_6, 2)
```

#### 8.Write a function that takes a list of words (containing duplicates) (i.e. words=['table','chair','desk','table','table','chair']) and returns a list of words (with no duplicates) sorted by decreasing frequency. E.g. if the input list contained 10 instances of the word table and 9 instances of the word chair, then table would appear before chair in the output list. You should use lambda in the sorted( ) function.

```python
words_8=['table','chair','desk','table','table','chair']
```

```python
def list_sorter(list_8):
    fq = FreqDist(list_8)
    freq = dict(sorted(fq.items(),key = lambda x: x[1], reverse = True))
    return list(freq.keys())
```

```python
list_sorter(words_8)
```

Write a function that takes a text (e.g. text3 from nltk.book) and a vocabulary (e.g. nltk.corpus.words.words()) as its arguments and returns the set of words that appear in the text but not in the vocabulary. Both arguments can be represented as lists of strings. Can you do this in a single line, using set.difference()?

```python
text_9 = nltk.book.text3
voc_9 = nltk.corpus.words.words()
```

```python
def text_diff_finder(text,voc):
    return set(text).difference(voc)
 
text_diff_finder(text_9,voc_9)
```

#### 10. Choose your own webpage (in html format) and output the 20 most common words in this web page. You should use w.lower() to normalize the text. Please get rid of stop words, numbers, and punctuations. Please define a function and call that function.

```python
url = 'https://www.theguardian.com/music/2021/jan/01/mf-doom-a-hip-hop-genius-who-built-his-own-universe-of-poetry'
html = request.urlopen(url).read().decode('utf8')
text = BeautifulSoup(html,'html.parser').get_text()
fd = nltk.FreqDist(word.lower() for word in nltk.word_tokenize(text) if word.isalpha() and word not in stopwords.words('english'))
[word.lower() for (word, _) in fd.most_common(20)]
```
